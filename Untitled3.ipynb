```python
# Import required libraries
from transformers import pipeline, AutoTokenizer, AutoModelForSeq2SeqLM

# Load pre-trained summarization model and tokenizer
tokenizer = AutoTokenizer.from_pretrained('facebook/bart-large-cnn')
model = AutoModelForSeq2SeqLM.from_pretrained('facebook/bart-large-cnn')

# Initialize a summarization pipeline
summarizer = pipeline("summarization", model=model, tokenizer=tokenizer)

# Define a function for summarization
def summarize_text(text, max_length=150, min_length=30, do_sample=False):
    """
    Summarizes the input text using the loaded model.

    Parameters:
    - text: The input text to summarize.
    - max_length: Maximum length of the summary.
    - min_length: Minimum length of the summary.
    - do_sample: Whether to use sampling for text generation.

    Returns:
    - A string containing the summarized text.
    """
    # Check if the text is too long; split into chunks if necessary
    if len(text) > 1024:  # Model-specific token limit (adjust for your model)
        chunks = split_into_chunks(text, max_chunk_length=1024)
        summaries = [summarizer(chunk, max_length=max_length, min_length=min_length, do_sample=do_sample)[0]["summary_text"] for chunk in chunks]
        return " ".join(summaries)
    else:
        summary = summarizer(text, max_length=max_length, min_length=min_length, do_sample=do_sample)
        return summary[0]["summary_text"]

# Define a helper function for splitting long texts (if needed)
def split_into_chunks(text, max_chunk_length):
    """
    Splits the input text into smaller chunks based on the model's token limit.
    """
    words = text.split()
    chunks = []
    current_chunk = []
    current_length = 0

    for word in words:
        if current_length + len(word) + 1 > max_chunk_length:
            chunks.append(" ".join(current_chunk))
            current_chunk = []
            current_length = 0
        current_chunk.append(word)
        current_length += len(word) + 1

    if current_chunk:
        chunks.append(" ".join(current_chunk))

    return chunks

# Use the pipeline
input_text = "Your long text to summarize goes here."
summary = summarize_text(input_text)
print("Summarized Text:", summary)
```

```python
import requests
import json

# Replace with your OpenWeatherMap API key
API_KEY = "d4e06f0bd751bf478fb587be3c749d61"
BASE_URL = "https://api.openweathermap.org/data/2.5/weather"

def fetch_weather_data(city_name, api_key):
    params = {
        "q": city_name,
        "appid": api_key,
        "units": "metric"  # Change to "imperial" for Fahrenheit
    }
    response = requests.get(BASE_URL, params=params)
    if response.status_code == 200:
        return response.json()
    else:
        print(f"Error: Unable to fetch data for {city_name}. Status code: {response.status_code}")
        return None

def save_to_json(data, filename):
    """
    Save the given data to a JSON file.

    Parameters:
    - data: The data to save.
    - filename: The name of the JSON file.
    """
    with open(filename, "w") as file:
        json.dump(data, file, indent=4)
    print(f"Data saved to {filename}")

# Main function
if __name__ == "__main__":
    city = input("Enter the city name: ")
    weather_data = fetch_weather_data(city, API_KEY)
    if weather_data:
        save_to_json(weather_data, f"{city}_weather.json")
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Target URL
url = "https://www.bbc.com/news"

response = requests.get(url)

if response.status_code == 200:
    soup = BeautifulSoup(response.content, "html.parser")

    # Extract headlines
    headlines = []
    for headline in soup.find_all("h3"):  # tag based on the website structure
        text = headline.get_text(strip=True)
        if text:
            headlines.append(text)

    # Save headlines to a DataFrame
    df = pd.DataFrame(headlines, columns=["Headline"])

    # Save to CSV
    df.to_csv("news_headlines.csv", index=False)
    print("Headlines saved to news_headlines.csv")
else:
    print(f"Error: Unable to fetch the webpage. Status Code: {response.status_code}")
```
